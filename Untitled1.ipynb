```python
import pandas as pd

# Load your dataset
df = pd.read_csv('final_complaints_updated.csv')

# Optional: Check which columns are categorical
print(df.dtypes)

# Convert categorical features to numeric using one-hot encoding
df_encoded = pd.get_dummies(df[['predicted_cost', 'area', 'type', 'department', 'actual_cost']], drop_first=True)

# Calculate correlation with 'actual_cost'
correlation_matrix = df_encoded.corr()

# Extract only the correlation of features with 'actual_cost'
actual_cost_corr = correlation_matrix['actual_cost'].sort_values(ascending=False)

# Print the result
print("Correlation of features with actual_cost:")
print(actual_cost_corr)
```

```python
import pandas as pd
import numpy as np

# Load your DataFrame from the existing CSV
df = pd.read_csv('final_complaints_updated.csv')

# Calculate the cost difference
df["cost_difference"] = np.abs(df["actual_cost"] - df["predicted_cost"])

# Define a threshold for anomaly detection, for example 30%
threshold = 0.3  # 30%

# Calculate the cost difference percentage
df["cost_diff_percent"] = df["cost_difference"] / df["predicted_cost"]

# Create the actual_cost_anomaly column based on the threshold
df["actual_cost_anomaly"] = df["cost_diff_percent"] > threshold

# Save the updated DataFrame back to the CSV file
df.to_csv('final_complaints_updated.csv', index=False)

# Print to confirm the update
print(df[["complaint_id", "actual_cost", "predicted_cost", "cost_difference", "cost_diff_percent", "actual_cost_anomaly"]].head())
```

```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Sample Data Generation
complaint_data = []

# Areas, Issues, and Departments
areas = ["Koregaon Park", "Shivajinagar", "Baner", "Kothrud", "Hadapsar", "Deccan", "Pimpri-Chinchwad", "Viman Nagar", "Aundh", "Warje"]
issues = ["Garbage Issue", "Electricity Issue", "Illegal Parking", "Tree Falling", "Road Damage", "Animal Nuisance", "Streetlight Issue", "Noise Pollution", "Water Issue", "Sewage Problem"]
departments = {
    "Garbage Issue": "Municipal Waste Management Department",
    "Electricity Issue": "Electricity Distribution Department",
    "Illegal Parking": "Traffic and Transportation Department",
    "Tree Falling": "Urban Forestry and Landscaping Department",
    "Road Damage": "Public Works Department",
    "Animal Nuisance": "Animal Control and Welfare Department",
    "Streetlight Issue": "Urban Lighting and Infrastructure Department",
    "Noise Pollution": "Environmental Protection Department",
    "Water Issue": "Water Supply and Sanitation Department",
    "Sewage Problem": "Sewerage and Drainage Department"
}

# Simulate 200 complaints
for i in range(1, 201):
    complaint_id = f"CID{i:03}"
    complaint_text = random.choice(issues)
    area = random.choice(areas)
    issue_type = complaint_text
    filing_date = datetime(2023, random.randint(1, 12), random.randint(1, 28)).strftime('%Y-%m-%d')
    department = departments[issue_type]
    predicted_priority = random.choice(["High", "Medium", "Low"])
    status = random.choice(["Resolved", "Pending"])
    resolved_days = random.randint(1, 15) if status == "Resolved" else None
    resolution_date = (datetime.strptime(filing_date, '%Y-%m-%d') + timedelta(days=resolved_days)) if resolved_days else None
    rating = random.randint(1, 5) if status == "Resolved" else None
    feedback_text = random.choice(["Excellent", "Good", "Average", "Poor"]) if status == "Resolved" else None
    
    # Generate cost data
    predicted_cost = random.randint(1000, 5000)
    actual_cost = predicted_cost + random.randint(-1000, 1000)  # To simulate anomalies
    actual_cost_anomaly = "Yes" if actual_cost != predicted_cost else "No"
    
    # Variance flag
    variance_flag = "Yes" if actual_cost_anomaly == "Yes" else "No"
    
    complaint_data.append([
        complaint_id, complaint_text, area, issue_type, filing_date, department, predicted_priority, status, resolved_days,
        resolution_date.strftime('%Y-%m-%d') if resolution_date else None, rating, feedback_text, actual_cost, predicted_cost,
        actual_cost_anomaly, variance_flag
    ])

# Create DataFrame
columns = [
    "complaint_id", "complaint_text", "area", "type", "filing_date", "department", "predicted_priority", "status", "resolved_days",
    "resolution_date", "rating", "feedback_text", "actual_cost", "predicted_cost", "actual_cost_anomaly", "variance_flag"
]

df_complaints = pd.DataFrame(complaint_data, columns=columns)

# Save to CSV
df_complaints.to_csv('complaints_data_200.csv', index=False)
```
